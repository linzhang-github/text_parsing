```python
with open('sample.txt', 'r') as f: 
    text = f.read()
```

```python
import nltk
import string 
from nltk.corpus import stopwords 

# remove puctuation in the text 
tokens = nltk.wordpunct_tokenize(text) 
words_rm_punc = [w.lower() for w in tokens if w.isalpha()] 
numbers = [w.lower() for w in tokens if w.isdigit()]  
stop_words = stopwords.words('english')
words_rm_stopwds = [w.lower() for w in words_rm_punc if not w in stop_words]


total_character = len(tokens)
count_character_wt_punc = len(words_rm_punc)
count_unique_character_wt_punc = len(set(words_rm_punc))
count_digit = len(numbers) 
count_character_wt_stopwds = len(words_rm_stopwds)

print(' total_character:', total_character,
    '\n','total_character_without_punctuation:', count_character_wt_punc,
     '\n', "total_unique_character_without_punctuation:", count_character_wt_punc,
     '\n', 'total_character_without_stopwords:', count_character_wt_stopwds, 
          '\n', 'total_digit:', count_digit, )
```

## Top 20 most frequent single words 

```python
from nltk.probability import FreqDist 
import matplotlib.pyplot as plt
%matplotlib inline  
fdist = FreqDist(words_rm_stopwds) 
fdist.most_common(20)
```

**Frequency of uni-gram**

```python
fdist.plot(20) 
```

## Top 10 or 20 most frequent bi-grams words in the text

**Frequency of Bigrams**

```python
bgs = nltk.bigrams(words_rm_stopwds)
nltk.FreqDist(bgs).most_common(20)
```

```python
bgs = nltk.bigrams(words_rm_stopwds)
nltk.FreqDist(bgs).plot(10)
```
